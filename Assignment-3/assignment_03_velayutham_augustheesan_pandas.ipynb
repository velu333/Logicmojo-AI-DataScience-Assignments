#### 10. How many Counties don't have the word "County" in their name?

```python
"""Count how many Counties do not have the word 'County' in their name"""

import pandas as pd
from pathlib import Path
from input_utils import load_from_file

# Data file path
DATA_DIR = Path("sample_data")
FILE_PATH = DATA_DIR / "population_by_county.csv"


def main() -> None:
    """Read CSV and display the number of counties without the word 'County'."""
    try:
        # Load data
        population_df = load_from_file(FILE_PATH)

        # Count counties without "County"
        no_county_count = (
            ~population_df["County"].str.contains("County", case=False, na=False)
        ).sum()

        # Display result
        print(f"The count of counties without the word 'County' is: {no_county_count}")

    except (FileNotFoundError, ValueError) as e:
        print(f"An error has occurred: {e}")


if __name__ == "__main__":
    main()
```

#### 11. Add a column that calculates the percent change between the 2010 Census and the 2017 Population Estimate

```python
"""Add a column to calculate percent change between 2010 Census and 2017 estimate."""

import numpy as np
import pandas as pd
from pathlib import Path
from input_utils import load_from_file

# Data file path
DATA_DIR = Path("sample_data")
FILE_PATH = DATA_DIR / "population_by_county.csv"


def main() -> None:
    """Load data, calculate percent change, and display the updated DataFrame."""
    try:
        # Load data
        population_df = load_from_file(FILE_PATH)

        # Calculate the percent change
        population_df["Percent Change"] = (
            (population_df["2017PopEstimate"] - population_df["2010Census"])
            / population_df["2010Census"]
        ) * 100

        # Handle infinite and NaN values
        population_df["Percent Change"] = (
            population_df["Percent Change"]
            .replace([np.inf, -np.inf], np.nan)
        )

        # Round values for readability
        population_df["Percent Change"] = population_df["Percent Change"].round(2)

        # Display the updated DataFrame
        print("\nDataFrame with the newly added column:\n")
        print(population_df.head(5).to_string())

    except (FileNotFoundError, ValueError, KeyError) as e:
        print(f"An error has occurred: {e}")


if __name__ == "__main__":
    main()
```

#### 12. What States have the highest estimated percent change between the 2010 Census and the 2017 Population Estimate?

**NOTE:** This will take several lines of code, as it requires a recalculation of Percent Change.

```python
"""States that have the highest estimated percent change between the 2010 Census and the 2017 Population Estimate."""

import pandas as pd
import numpy as np
from pathlib import Path
from input_utils import load_from_file

# Define the data file path
DATA_DIR = Path("sample_data")
FILE_PATH = DATA_DIR / "population_by_county.csv"


def main() -> None:
    """Find states with the highest percent change between 2010 Census and 2017 Population Estimate."""
    try:
        # Read the data into a Pandas DataFrame
        population_df = load_from_file(FILE_PATH)

        # Group by state and sum populations
        state_totals = population_df.groupby("State")[["2010Census", "2017PopEstimate"]].sum()

        # Calculate percent change
        state_totals["Percent Change"] = (
            (state_totals["2017PopEstimate"] - state_totals["2010Census"])
            / state_totals["2010Census"]
        ) * 100

        # Sort states by percent change
        top_states = state_totals.sort_values("Percent Change", ascending=False).head(5)
        top_states_reset = top_states.reset_index()

        # Display results
        print("\nTop 5 States by Percent Change (2010-2017):\n")
        print(top_states_reset[["State", "Percent Change"]].round(2).to_string(index=False))

    except (FileNotFoundError, ValueError, KeyError) as e:
        print(f"An error has occurred: {e}")


if __name__ == "__main__":
    main()
```
